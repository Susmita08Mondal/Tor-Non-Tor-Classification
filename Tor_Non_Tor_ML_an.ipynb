```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data[' Source Port'] = le.fit_transform(data[' Source Port'])
```

```python
dp = preprocessing.LabelEncoder()
data[' Destination Port'] = dp.fit_transform(data[' Destination Port'])
```

```python
tp = preprocessing.LabelEncoder()
data['label'] = tp.fit_transform(data['label'])
```

```python
data.head()
```

```python
data_source=data.drop(columns=['Source IP',' Destination IP', ' Source Port',' Destination Port'])
```

```python
data_source.columns
```

```python
flags_ohe = data_source
categorical_columns = [' Protocol']
for col in categorical_columns:
    col_ohe = pd.get_dummies(data_source[col], prefix=col)
    flags_ohe = pd.concat((flags_ohe, col_ohe), axis=1).drop(col, axis=1)
```

```python
import numpy as np
flags_ohe=flags_ohe.fillna(0)
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
flags_ohe['label']= label_encoder.fit_transform(flags_ohe['label'])
  
flags_ohe['label'].unique()
```

```python
for i in flags_ohe.columns:
  flags_ohe=flags_ohe[(flags_ohe[i]!='inf')]
```

```python
flags_ohe.drop(flags_ohe.tail(6).index,
        inplace = True)
```

```python
from sklearn.model_selection import train_test_split

#train, test = train_test_split(flags_ohe, test_size = 0.1)
df = flags_ohe.fillna(0)
train_x, test_x, train_y, test_y = train_test_split(df[df.columns.drop(['label'])], df['label'],test_size=0.1)
```

```python

# example of a standardization
from numpy import asarray
from sklearn.preprocessing import StandardScaler
train_x = train_x.astype(np.float32)
col = train_x.columns
scaler = StandardScaler()
train_x = scaler.fit_transform(train_x)
test_x = scaler.transform(test_x)
```

```python
from sklearn.linear_model import SGDClassifier, SGDRegressor,LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import random
import itertools
```

```python

train_x = pd.DataFrame(train_x, columns = col)
test_x = pd.DataFrame(test_x, columns = col)
```

Logistic Regression

```python
basicmodel = LogisticRegression()
basicmodel = basicmodel.fit(train_x, train_y)
preds1 = basicmodel.predict(test_x)
acc1=accuracy_score(test_y, preds1)
print("Accuracy is {}".format(acc1*100))
print(classification_report(test_y, preds1))
print(confusion_matrix(test_y, preds1))
```

Random Forest

```python
basicmodel = RandomForestClassifier()
basicmodel = basicmodel.fit(train_x, train_y)
preds1 = basicmodel.predict(test_x)
acc1=accuracy_score(test_y, preds1)
print("Accuracy is {}".format(acc1*100))
print(classification_report(test_y, preds1))
print(confusion_matrix(test_y, preds1))
```

Gradient Boosting Machines

```python
  basicmodel = GradientBoostingClassifier()
basicmodel = basicmodel.fit(train_x, train_y)
preds1 = basicmodel.predict(test_x)
acc1=accuracy_score(test_y, preds1)
print("Accuracy is {}".format(acc1*100))
print(classification_report(test_y, preds1))
print(confusion_matrix(test_y, preds1))
```

Stochastic Gradient Descent Classifier

```python
basicmodel= SGDClassifier(loss='modified_huber', max_iter=5, random_state=0, shuffle=True)
basicmodel = basicmodel.fit(train_x, train_y)
preds1 = basicmodel.predict(test_x)
acc1=accuracy_score(test_y, preds1)
print("Accuracy is {}".format(acc1*100))
print(classification_report(test_y, preds1))
print(confusion_matrix(test_y, preds1))
```

```python
from sklearn import svm
basicmodel = svm.SVC(kernel='sigmoid')
basicmodel.fit(train_x, train_y)
preds1 = basicmodel.predict(test_x)
acc1=accuracy_score(test_y, preds1)
print("Accuracy is {}".format(acc1*100))
print(classification_report(test_y, preds1))
print(confusion_matrix(test_y, preds1))
```
